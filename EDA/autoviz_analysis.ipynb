```python
import os
from autoviz.AutoViz_Class import AutoViz_Class

# ✅ Create output folder manually (safe even if it already exists)
output_dir = "autoviz_results"
os.makedirs(output_dir, exist_ok=True)

# ✅ Initialize AutoViz
AV = AutoViz_Class()

# ✅ Path to your CSV
filename = r"C:\Users\Home\PycharmProjects\DSC-project\train.csv"
target_variable = "Listening_Time_minutes"

# ✅ Run AutoViz and save plots in the folder
dft = AV.AutoViz(
    filename=filename,
    sep=",",
    depVar=target_variable,
    dfte=None,
    header=0,
    verbose=1,
    lowess=False,
    chart_format="html",   # use "html" for easy viewing in browser
    max_rows_analyzed=800000,
    max_cols_analyzed=30,
    save_plot_dir=output_dir  # this makes AutoViz save all plots
)

print(f"\n✅ All plots saved in: {os.path.abspath(output_dir)}")
```
